```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.title('Weather')
plt.plot(x,y,color='c',linewidth=2, linestyle='-.')
plt.show()
```

```python
plt.plot(x,y,color='r',linewidth=8, linestyle=':')
plt.show()
```

```python
plt.plot(x,y,color='green',linewidth=5, linestyle='-.')
```
